```python
job_description = docx2txt.process('AnswerSheet.docx')
Result =docx2txt.process('modelAnswer.docx')
```

```python
print(Result)
```

```python
content =(job_description,Result)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
matrix = cv.fit_transform(content)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix=cosine_similarity(matrix)
```

```python
print(similarity_matrix)
```

```python
print('Result Score  :'+str(similarity_matrix[1][0]*100)+'%')
```
